```python
import matplotlib.pyplot as plt
import numpy as np

# Funksjon som tegner en spirograf
def draw_spirograph(R,r,l):
    # Lager figur og akser
    fig, ax = plt.subplots()        
    
    # Lager vinkelverdier, r / np.gcd(R, r) regner ut hvor mange runder for  f hel figur 
    theta = np.linspace(0, 2 * np.pi* r / np.gcd(R, r) , 1000) 

    # Bruker vinkelverdiene til  lage x- og y- verdier
    x = (R - r) * np.cos(theta) + l * r * np.cos((R - r) * theta / r)
    y = (R - r) * np.sin(theta) - l * r * np.sin((R - r) * theta / r)
 
    # Plotter spirografen
    ax.plot(x, y,linewidth=1)
    
    # Fjerne aksene
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Fjerne akseetikettene
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    
    # Setter enhetene p aksene like store for  vise sirkelen som en sirkel
    ax.set_aspect('equal')

    # viser plottet
    plt.show()

# Eksempler p parametere for spirografen
R = 180  # Radius til den stasjonre sirkelen
r = 75   # Radius til den rullende sirkelen
l = 0.5  # Hvor langt ute p sirkelen setter du pennen?

# tegner spirografen
draw_spirograph(R,r,l)
```

```python
# gjr spirgografen interaktiv
from ipywidgets import interact, widgets, Layout

interact(draw_spirograph, 
         R=widgets.IntSlider(min=0, max=200, step=1, value=150,layout=Layout(width='500px')), 
         r=widgets.IntSlider(min=0, max=200, step=1, value=70,layout=Layout(width='500px')), 
         l=widgets.FloatSlider(min=0, max=1, step=0.05, value=0.5,layout=Layout(width='500px')));        
```
